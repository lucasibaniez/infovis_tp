```python
import pandas as pd
```

```python
data = "datos_nomivac_covid19.csv"
```

```python
# Leemos Dataframe
df = pd.read_csv(data, dtype= {'jurisdiccion_residencia_id' : object, 
                                                      'depto_residencia_id'        : object, 
                                                      'jurisdiccion_aplicacion_id' : object, 
                                                      'depto_aplicacion_id'        : object})
```

```python
df.head()
```

```python
# Filtramos Vacunados por cada dosis
df_1dosis = df[(df.orden_dosis == 1)]
df_2dosis = df[(df.orden_dosis == 2)]
```

```python
df_1dosis.head()
```

```python
df_2dosis.head()
```

```python
# Agrupamos por dosis y jurisdiccion
df_agrupado_1dosis = df_1dosis.groupby(['jurisdiccion_aplicacion']).size().reset_index(name='counts')
df_agrupado_2dosis = df_2dosis.groupby(['jurisdiccion_aplicacion']).size().reset_index(name='counts')
```

```python
df_agrupado_2dosis
```

```python
df_agrupado_1dosis
```

```python
# Cantidad de Habitantes por jurisdiccion
# Fuente: 
poblacion = {
    "Buenos Aires": 17709598,
    "CABA": 3078836,
    "Catamarca": 418991,
    "Chaco": 1216247,
    "Chubut": 629181,
    "Corrientes": 1130320,
    "Córdoba": 3798261,
    "Entre Ríos": 1398510,
    "Formosa": 610019,
    "Jujuy": 779212,
    "La Pampa": 361394,
    "La Rioja": 398648,
    "Mendoza": 2010363,
    "Misiones": 1274992,
    "Neuquén": 672461,
    "Río Negro": 757052,
    "Salta": 1441988,
    "San Juan": 789489,
    "San Luis": 514610,
    "Santa Cruz": 374756,
    "Santa Fe": 3563390,
    "Santiago del Estero": 988245,
    "Tierra del Fuego": 177697,
    "Tucumán": 1714487
}

poblacion_df = pd.DataFrame(poblacion.items(), columns=['jurisdiccion', 'habitantes'])
poblacion_df
```

```python
# Unimos 
densidad_1dosis = pd.merge(df_agrupado_1dosis, poblacion_df, how='left', left_on=['jurisdiccion_aplicacion'], right_on=['jurisdiccion'])
densidad_1dosis["porcentaje_vacunados"] = densidad_1dosis.counts * 100 / densidad_1dosis.habitantes
densidad_1dosis = densidad_1dosis.drop(columns=['counts', 'jurisdiccion_aplicacion', 'habitantes'])
densidad_1dosis
```

```python
# Unimos 
densidad_2dosis = pd.merge(df_agrupado_2dosis, poblacion_df, how='left', left_on=['jurisdiccion_aplicacion'], right_on=['jurisdiccion'])
densidad_2dosis["porcentaje_vacunados"] = densidad_2dosis.counts * 100 / densidad_2dosis.habitantes
densidad_2dosis = densidad_2dosis.drop(columns=['counts', 'jurisdiccion_aplicacion', 'habitantes'])
densidad_2dosis
```

```python
# Exportamos json
densidad_1dosis.to_json('densidad1dosis.json', orient = 'records')
densidad_2dosis.to_json('densidad2dosis.json', orient = 'records')
```
