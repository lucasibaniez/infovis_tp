```python
# Importamos librerias
import pandas as pd
```

```python
# Variables a utilizar
data = "datos_nomivac_covid19.csv"
# Leemos Dataframe
df = pd.read_csv(data, dtype= {'jurisdiccion_residencia_id' : object, 
                                                      'depto_residencia_id'        : object, 
                                                      'jurisdiccion_aplicacion_id' : object, 
                                                      'depto_aplicacion_id'        : object})
```

```python
# Filtramos Vacunados por cada dosis
df_1dosis = df[(df.orden_dosis == 1)]
```

```python
# 2da Dosis
df_2dosis = df[(df.orden_dosis == 2)]
df_2dosis
```

```python
# Parseamos 1ra Dosis
df_1dosis['fecha_aplicacion'] = pd.to_datetime(df_1dosis['fecha_aplicacion'])
df_1dosis
#df_1dosis["mes"] = df_1dosis["fecha_aplicacion"].str.split("-")[1]
#df_1dosis
```

```python
# Parseamos 2da Dosis
df_2dosis['fecha_aplicacion'] = pd.to_datetime(df_2dosis['fecha_aplicacion'])
df_2dosis
```

```python
# 1ra Dosis: Generamos columna de mes numerico
df_1dosis['mes'] = df_1dosis['fecha_aplicacion'].dt.month
df_1dosis
```

```python
# 2da Dosis: Generamos columna de mes numerico
df_2dosis['mes'] = df_2dosis['fecha_aplicacion'].dt.month
df_2dosis
```

```python
# 1ra Dosis: Agrupamos por vacuna y mes
df_agrupado = df_1dosis.groupby(['vacuna', 'mes']).size().reset_index(name='counts')
df_agrupado = df_agrupado.sort_values(by=['mes'])
df_agrupado
```

```python
# 2da Dosis: Agrupamos por vacuna y mes
df_agrupado2 = df_2dosis.groupby(['vacuna', 'mes']).size().reset_index(name='counts')
df_agrupado2 = df_agrupado2.sort_values(by=['mes'])
df_agrupado2
```

```python
# 1ra Dosis: Corregimos diciembre 2020
df_agrupado.loc[37, "mes"] = 0
df_agrupado = df_agrupado.sort_values(by=['mes'])
df_agrupado
```

```python
# 1ra Dosis: Corregimos diciembre 2020
df_agrupado2.loc[33, "mes"] = 0
df_agrupado2 = df_agrupado2.sort_values(by=['mes'])
df_agrupado2
```

```python
vacunas = list(df_agrupado.vacuna.unique())
vacunas
```

```python
# 1ra Dosis: Generamos registros vacios
for vac in vacunas:
    df_vac = df_agrupado[(df_agrupado.vacuna == vac)]
    print(vac)
    for x in range(0,10):
        t = df_vac[(df_vac.mes == x)]
        if t.empty:
            df2 = {'vacuna': vac, 'mes': x, 'counts': 0}
            df_agrupado = df_agrupado.append(df2, ignore_index = True)

df_agrupado = df_agrupado.sort_values(by=['mes'])
df_agrupado
```

```python
# 2da Dosis: Generamos registros vacios
for vac in vacunas:
    df_vac = df_agrupado2[(df_agrupado2.vacuna == vac)]
    for x in range(0,10):
        t = df_vac[(df_vac.mes == x)]
        if t.empty:
            df2 = {'vacuna': vac, 'mes': x, 'counts': 0}
            df_agrupado2 = df_agrupado2.append(df2, ignore_index = True)

df_agrupado2 = df_agrupado2.sort_values(by=['mes'])
df_agrupado2
```

```python
# 1ra Dosis: Convertimos df a diccionario
lista = df_agrupado.to_dict('records')
lista
```

```python
lista2 = df_agrupado2.to_dict('records')
lista2
```

```python
# Obtenemos acumulados: 1ra Dosis
for vac in vacunas:
    acum = 0
    for mes in range(0,10):
        x = next(item for item in lista if item["vacuna"] == vac and item["mes"] == mes)
        acum+=x["counts"]
        x["counts"]=acum

lista
```

```python
# Obtenemos acumulados: 2da dosis
for vac in vacunas:
    acum = 0
    for mes in range(0,10):
        x = next(item for item in lista2 if item["vacuna"] == vac and item["mes"] == mes)
        acum+=x["counts"]
        x["counts"]=acum

lista
```

```python
# Generamos salida: 1ra dosis
salida = {}

for mes in range(0,10):
    salida[mes] = []
    for vac in vacunas:
        x = next(item for item in lista if item["vacuna"] == vac and item["mes"] == mes)
        salida[mes].append({"vacuna":vac, "cantidad":x["counts"]})

salida
```

```python
# Generamos salida: 2da dosis
salida2 = {}

for mes in range(0,10):
    salida2[mes] = []
    for vac in vacunas:
        x = next(item for item in lista2 if item["vacuna"] == vac and item["mes"] == mes)
        salida[mes].append({"vacuna":vac, "cantidad":x["counts"]})

salida2
```

```python
# Exportamos
import json

file = open("vacunas_acumuladas.json", "w")
file.write(json.dumps(salida))
file.close()
```

```python
file = open("vacunas_acumuladas2.json", "w")
file.write(json.dumps(salida2))
file.close()
```
