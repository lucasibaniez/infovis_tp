```python
# Importamos librerias
import pandas as pd
```

```python
# Variables a utilizar
data = "datos_nomivac_covid19.csv"
```

```python
# Leemos Dataframe
df = pd.read_csv(data, dtype= {'jurisdiccion_residencia_id' : object, 
                                                      'depto_residencia_id'        : object, 
                                                      'jurisdiccion_aplicacion_id' : object, 
                                                      'depto_aplicacion_id'        : object})
df.head()
```

```python
# Obtenemos los vacunados en el Chaco
df_vacunados_chaco = df[(df.jurisdiccion_aplicacion == "Chaco")]
df_vacunados_chaco.head()
```

```python
# Quitamos registros cuya jurisrisdiccioón de Residencia sea Chaco
df_vacunados_chaco_otra_jur = df_vacunados_chaco[(df_vacunados_chaco.jurisdiccion_residencia != "Chaco")]
df_vacunados_chaco_otra_jur
```

```python
# Agrupamos por jurisdiccion de residencia
df_agrupado = df_vacunados_chaco_otra_jur.groupby(['jurisdiccion_residencia']).size().reset_index(name='counts')
df_agrupado
```

```python
# Exportamos
df_agrupado.to_json('foraneos_chaco.json', orient = 'records')
```

```python
# Agrupamos por jurisdiccion_residencia y jurisdiccion_aplicacion
df_agrupado_pais = df.groupby(['jurisdiccion_residencia', 'jurisdiccion_aplicacion']).size().reset_index(name='counts').sort_values(by=['jurisdiccion_aplicacion', 'counts'])
df_agrupado_pais
```

```python
# Jurisdicciones
jurisdicciones = list(df_agrupado_pais.jurisdiccion_aplicacion.unique())
jurisdicciones
```

```python
data_graph = []
for jur in jurisdicciones:
    data = {}
    data["jurisdiccion"] = jur
    df_temp = df_agrupado_pais[(df_agrupado_pais.jurisdiccion_aplicacion == jur)]
    data["foraneos"] = int(df_temp['counts'].sum())
    data["particion"] = []
    jur_foraneas = list(df_temp.jurisdiccion_residencia.unique())
    for jur_res in jur_foraneas:
        if jur_res == jur:
            continue
        df_jur_res = df_temp[(df_temp.jurisdiccion_residencia == jur_res)]
        data["particion"].append({"jurisdiccion": jur_res, "counts": int(df_jur_res['counts'].sum())})
    data_graph.append(data)

data_graph
```

```python
# Exportamos
import json

file = open("foraneos_pais.json", "w")
file.write(json.dumps(data_graph))
file.close()
```
