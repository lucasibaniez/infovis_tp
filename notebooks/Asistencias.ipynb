```python
import pandas as pd
```

```python
FECHA_INFORME = '20210727'
```

## Tratamos archivo del zoom

```python
df_zoom = pd.read_csv('%s.csv' % (FECHA_INFORME))
# Renombramos las columnas
rename_col = {
    'Nombre (nombre original)': 'zoom_nombre_completo',
    'E-mail del usuario': 'zoom_email',
    'Hora para unirse': 'zoom_hora_inicio',
    'Hora para salir': 'zoom_hora_inicio',
    'Duración (minutos)': 'zoom_duracion',
    'Invitado': 'zoom_invitado'
}

df_zoom = df_zoom.rename(columns=rename_col)
df_zoom
```

## Tratamos Lista de Alumnos

```python
df_cv = pd.read_csv('ListaAlumnos02.csv')
rename_col = {
    'Nombre': 'cv_nombre',
    'Apellido(s)': 'cv_apellido',
    'Nombre de usuario': 'cv_username',
    'Dirección de correo': 'cv_email',
    'Ciudad': 'cv_ciudad'
}
df_cv = df_cv.rename(columns=rename_col)
df_cv
```

## Cruzamos información

```python
# Obtenemos de la lista del Aula Virtual los que asistieron
salida = pd.merge(df_cv, df_zoom, how='left', left_on=['cv_email'], right_on=['zoom_email'])
salida = salida.drop_duplicates(subset=['cv_username'])
salida
```

```python
# Verificamos emails duplicados
duplicateRowsDF = salida[salida.duplicated(['cv_username'], keep=False)]
duplicateRowsDF
```

## Analisis de datos

```python
salida.zoom_email.isnull().sum()
```

```python
salida.zoom_email.notnull().sum()
```

```python
salidaCoincidentes = salida[salida['zoom_email'].notna()]
salidaCoincidentes
```

```python
# Verificamos emails duplicados
duplicateRowsDF = salidaCoincidentes[salidaCoincidentes.duplicated(['cv_username'], keep=False)]
duplicateRowsDF
```

```python
# Exportamos
salidaCoincidentes.to_excel('%s AsistenciaZoom.xlsx' % (FECHA_INFORME), index=False)
```

## Cruce de informacion 2

```python
salidaNoCoincidentes = pd.merge(df_cv, df_zoom, how='right', left_on=['cv_email'], right_on=['zoom_email'])
salidaNoCoincidentes = salidaNoCoincidentes[salidaNoCoincidentes['cv_email'].isna()]
salidaNoCoincidentes
```

```python
# Verificamos emails duplicados
duplicateRowsDF = salidaNoCoincidentes[salidaNoCoincidentes.duplicated(['zoom_email'], keep=False)]
duplicateRowsDF
```

```python
# Eliminamos Duplicados
salidaNoCoincidentes = salidaNoCoincidentes.drop_duplicates(subset=['zoom_email'])
duplicateRowsDF = salidaNoCoincidentes[salidaNoCoincidentes.duplicated(['zoom_email'], keep=False)]
duplicateRowsDF
```
